
Nestes casos, focamos somente no conteúdo que é preciso pra resolver cada problema:

* Calcular a soma total --> somente os valores do bd
* Calcular a media --> somente os valores do bd e o tamanho dele
* Calcular o top5 max --> somente os valores do bd ordenados
* Calcular o top5 min --> somente os valores do bd ordenados

Ou seja, somente precisamos utilizar os valores "valor" ou "preco" do bd.

# Pegar todos os valores 'valor' do bd:

```python
# percebendo que precisa só da chave 'preço', pegar todos os itens dos dicts
lista_valores = []

### for tradicional
# for pra acessar cada item do bd
for dic in bd:
    valor = dic['valor']
    lista_valores.append(valor)

print(lista_valores)
```

```python
### list comprehension
lista_valores = [dic['valor'] for dic in bd]
print(lista_valores)
```

```python
### lambda + alguma funcao (map)
# como usar funcao lambda: lambda parametro1, parametro2, ... : expressao
# map: aplica uma funcao em cada item da iteravel ---- map(funcao, iteravel)


# O map funciona assim: cada item do iterável é aplicado uma função
# Imagine que queremos uma funcão que retorna a key 'valor' do item da lista
# map -> bd[0] -> funcao(bd[0]) -> bd[0]['valor']
#         bd[1] -> funcao(bd[1]) -> bd[1]['valor']

# para definir uma função que retorna a chave valor dado um dicionario como parâmetro
def pegar_o_valor(dic):
    return dic['valor']

# AGORA convertendo pra uma função lambda
# lambda parametro1, parametro2, ... : expressao
lambda dic : dic['valor']

# aplicando na função map
lista_valores = list(map(lambda dic : dic['valor'], bd))
print(lista_valores)

# APLICAÇÃO DA FUNÇÃO LAMBDA -- EXEMPLO
# # exemplo 1 - convertendo todo o conteúdo de uma tupla para float
# tupla_str = ('1.0', '3.7', '5.4')
# tupla_float = tuple(map(float, tupla_str)) # função: float; coleção: tupla_str
# print(tupla_str, tupla_float)
```

# Calculos

Resolvendo as tasks com nossa lista_valores:

```python
#---------
# soma dos valores  ---> 'preco' em todos os itens da lista
#---------
soma_elementos = sum(lista_valores)
print('soma:', soma_elementos)
# média dos valores ---> 'preco' em todos os itens da lista + qtd de itens da lista
soma_elementos
qtd_elementos = len(lista_valores)
media_elementos = soma_elementos / qtd_elementos
print('media:', media_elementos)

#---------
# top 5 dos valores max ---> 'preco' em todos os itens da lista ordenados
#---------
elementos_ordenados = sorted(lista_valores, reverse=True)
# elementos_ordenados[0],el<ementos_ordenados[1], elementos_ordenados[2], ....
max5 = elementos_ordenados[:5]
print('top5 max:', max5)

#---------
# top 5 dos valores min ---> 'preco' em todos os itens da lista ordenados
#---------
min5 = elementos_ordenados[-5:]
print('top5 min:', min5)

#---------
# top 5 dos valores mediana ---> 'preco' em todos os itens da lista ordenados + qtd de itens da lista
#---------
# valor central: elementos ordenados e queremos o dado do meio -> qtd_elementos/2 
mediana = elementos_ordenados[len(elementos_ordenados)//2-2: len(elementos_ordenados)//2+3]

# top 5 dos valores média ---> 'preco' em todos os itens da lista + qtd de itens da lista
# calcular a media
media_elementos

# ordena esses valores em ordem crescente e pega os 5 primeiros em relação ao absoluto ou modulo de (num - media)
med5 = sorted(lista_valores, key = lambda valor: abs(valor - media_elementos))[:5]
print('top5 med:', med5)
```

```python
# DESAFIO: entender como funciona o sorted(key=lambda)
# passo a passo de como ordenar por uma key:

# resultado
print('resultado',sorted(lista_valores, key = lambda valor: abs(valor - media_elementos)))

# no fim das contas, o que acontece por trás do resultado é
print('\n0o passo - calcular media:',media_elementos)
print('1o passo - subtrair a media da lista:',list(map(lambda x: x-media_elementos,lista_valores)))
print('2o passo - aplicar módulo ou valor absoluto:',list(map(lambda x: abs(x-media_elementos),lista_valores)))
lista_valores_com_posicao = list(map(lambda x: (x[0],abs(x[1]-media_elementos)), enumerate(lista_valores)))
print('3o passo - gravar a posicao de cada item da lista:',lista_valores_com_posicao)
lista_ordenada_pelo_mais_proximos_da_media = sorted(lista_valores_com_posicao, key= lambda x: x[1])
print('4o passo - ordenar sobre os valores (ou seja, distancia da média):', lista_ordenada_pelo_mais_proximos_da_media)
posicoes_gravadas_ordenada_pelo_mais_proximos_da_media = [i[0] for i in lista_ordenada_pelo_mais_proximos_da_media]
print('5o passo - retornar os valores das posições:',[lista_valores[posicao_original] for posicao_original in posicoes_gravadas_ordenada_pelo_mais_proximos_da_media])
```

```python
# ---
# por categoria

# percebemos que todos os resultados passados foram feitos somente com a lista_valor,
# ou seja, eu só preciso aplicar a MESMA OPERAÇÃO mas para uma lista de precos daquela categoria!
# Entao vamor trocar a `lista_valors` por uma `lista_valor_por_categoria`

# Eu escolhi transformar esta variavel em um dicionário de listas, 
# onde a chave é a categoria e o valor são os precos da categoria

lista_valor_por_categoria = {}

# pegar as chaves: obter todas as categorias uma unica vez
lista_categorias = list(map(lambda item: item['categoria'], bd))
# loop for key in keys 
for key in lista_categorias:
    # se não tivermos esta chave, adicionamos no dict
    if key not in lista_valor_por_categoria.keys():
        lista_valor_por_categoria[key] = []

# list comprehension 
lista_valor_por_categoria = {k:[] for k in lista_categorias}

# set + list comprehension -- `set` é um objeto do python em que não há valores duplicados
categorias_unicas = set(lista_categorias)
lista_valor_por_categoria = {k:[] for k in categorias_unicas}

#
```
